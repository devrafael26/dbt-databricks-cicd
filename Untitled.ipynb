```python
from pyspark.sql import SparkSession
import pandas as pd

# Seu DataFrame pandas (já extraído do SQL Server)
df_ecom = pd.read_csv("ecommerce.csv")  # ou vindo do seu SQL Server direto

# Crie uma SparkSession com conexão ao Databricks
spark = SparkSession.builder \
    .appName("Upload local para Databricks") \
    .config("spark.databricks.service.address", "https://<SEU-HOST>.cloud.databricks.com") \
    .config("spark.databricks.token", "<SEU-TOKEN>") \
    .getOrCreate()

# Converte pandas para Spark DataFrame
df_spark = spark.createDataFrame(df_ecom)

# Salva como tabela no Databricks (em um catálogo/schema que você tenha permissão)
df_spark.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("default.ecommerce")
```
